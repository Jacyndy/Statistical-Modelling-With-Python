<h2>Set up Libraries and Set up API Requests</h2>

Send a request to CityBikes for the city of your choice. 

```python
#importing libaries
import requests
import json
import pandas as pd
print('set-up complete')
```

```python
#Make the API request and check if site truely exist with a success code of 200
# define the url
url = "https://api.citybik.es/v2/networks"

# define the response
response = requests.get(url)

# print status code to check if the api was called correctly
print(response.status_code)
```

```python
#check API response in json format
json_file = response.json()

json_file
```

```python
# check the keys in the json file for ease parsing
json_file.keys()
```

```python
#Extract all the city names
all_cities = [network["location"]["city"] for network in json_file["networks"]]

#To print the list of all the cities,sort in Alpha Order and set - to remove duplicate if exist
for city in sorted(set(all_cities)): 
    print(city)
```

```python
#Get the Total number of cities
print("The Total number of cities:", len(all_cities))
```

<h3>Question 1 </h3>
Send a request to CityBikes for the city of your choice.

```python
#Send a request to CityBikes for the city of your choice.
city_name = "Vancouver"
#Find the network for the selected city
network_id = None
for net in json_file["networks"]:
    if city_name.lower() in net["location"]["city"].lower():
        network_id = net["id"]
        break
        
else:
    print(f"No bike network found in {city_name}.")
```

<h3>Question 2 </h3>
Parse through the response to get the details you want for the bike stations in that city (latitude, longitude, number of bikes).

```python
#API call to get network details - ID and n
network_details = requests.get(f"https://api.citybik.es/v2/networks/{network_id}")
network_data = network_details.json()

# Extract station details
stations = network_data["network"]["stations"]

#Print the Stations in Selected city, 0-4 rows
print(f"Bike Stations in {city_name}:")
for station in stations[0:4]:  
    name = station["name"]
    lat = station["latitude"]
    lon = station["longitude"]
    bikes = station["free_bikes"]
print(f"Station: {name}, Location: ({lat}, {lon}), Available Bikes: {bikes}")
```

<h3>Question 3 </h3>
Put your parsed results into a DataFrame.

```python
# Create a DataFrame
df = pd.DataFrame([{ "Station Name": station["name"],
                    "Latitude": station["latitude"],
                    "Longitude": station["longitude"],
                    "Available Bikes": station["free_bikes"]
                   } 
                   for station in stations])
# Display the first few rows
print(df.head())
```
