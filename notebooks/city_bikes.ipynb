# CityBikes

Send a request to CityBikes for the city of your choice. 

```python
#importing libaries
import requests
import json
import pandas as pd
```

```python
# creating citybikes url variable

url = 'https://api.citybik.es/v2/networks'

#creating response variable
response = requests.get(url)
```

```python
#converting the response to a json file using .json()
response_json = response.json()
```

```python
#calling the json file
response_json
```

```python
#finding the keys in the json file. This returns networks as the only key. This is a dictionary.
response_json.keys()
```

```python
#checking to see what values networks key contain. This is a list of dictionary
networks = response_json['networks']
#see what is in the list of networks
list(networks[0])
```

```python
# check the length of the items in the list
len(networks)
```

```python
# using find, we found vancouver as the popular city in the data
# storing the url in the variable below 
network_id_url = 'https://api.citybik.es/v2/networks/mobibikes'
network_id = requests.get(network_id_url)
network_json = network_id.json()
```

```python
# to view the json
network_json['network']
```

```python
# check the length of the newtwork_id response
len(network_json)
```

```python
#  station is a dictionary
# we are interested in stations
stations = network_json['network']['stations']
```

```python
# stations is a list
#  length to see number of stations
len(stations) # 258 stations
```

```python
# convert the list to pd. json_normalize works with list of dictionaries
df_citybikes = pd.json_normalize(stations)
```

```python
# convert the dataframe to csv
df_citybikes.to_csv('city_bikes_df.csv') #save to the repo
```
